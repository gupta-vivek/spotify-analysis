### Inspecting data

```python
df.info()
```

#### Ideally the columns - 'streams', 'in_deezer_playlists', 'in_shazam_charts' should be of int type, but they are not. Let's check why they are not int

#### Column: streams

```python
df.loc[~df['streams'].str.isdigit(), 'streams']
```

#### There is one string value present in streams column. Replacing it with 0

```python
df.loc[~df['streams'].str.isdigit(), 'streams'] = '0'
df['streams'] = df['streams'].astype('int64')
```

#### Column: in_deezer_playlist

```python
df.loc[~df['in_deezer_playlists'].str.isdigit(), 'in_deezer_playlists']
```

#### Commas are present in the column. They should be removed.

```python
df['in_deezer_playlists'] = df['in_deezer_playlists'].str.replace(',', '')
df['in_deezer_playlists'] = df['in_deezer_playlists'].astype('int64')
```

#### Column: in_shazam_charts

```python
df['in_shazam_charts'] = df['in_shazam_charts'].fillna('').astype(str) # Replacing null values with empty string, so str.isdigit() function can be used
df.loc[~df['in_shazam_charts'].str.isdigit(), 'in_shazam_charts']
```

#### There are some missing values and commas present in the column. They should be removed

```python
df['in_shazam_charts'] = df['in_shazam_charts'].str.replace(',', '')
df.loc[~df['in_shazam_charts'].str.isdigit(), 'in_shazam_charts'] = '0' # Replacing missing values with zero
df['in_shazam_charts'] = df['in_shazam_charts'].astype('int64')
```
